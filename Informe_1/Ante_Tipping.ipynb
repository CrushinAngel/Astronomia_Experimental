```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

df = pd.read_excel(r'C:\Users\nicol\Documents\Astro_Exp\Radio 1\antdip_AE2025A.xlsx', sheet_name=0)
lndw = df['ln(dP)'].tolist()[0: -1]
neg_cosec_el = -1/np.sin(np.deg2rad(df["EL"][0: -1].tolist()))

coeffs = np.polyfit(neg_cosec_el, lndw, deg=1)
fit = np.poly1d(coeffs)


plt.figure(figsize = (15, 15))
plt.plot(neg_cosec_el, fit(neg_cosec_el), color = 'orange', label = 'Fit', ls = '--')
plt.plot(neg_cosec_el, lndw, color = 'magenta', label = 'Unfit')
plt.legend()
plt.ylabel('$ln(\Delta w)$')
plt.xlabel('$-cosec(el)$')
plt.title('Fit of $ln(\Delta w)$ v/s $-cosec(el)$ to find $\\tau_0$')
plt.text(0.01, 0.95, "$\\tau_0 = {0:.9f}$".format(coeffs[0]), fontsize=12, color='black', ha='left', va='top', transform=plt.gca().transAxes)
plt.show()

```
