```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

df = pd.read_excel(r'C:\Users\nicol\Documents\Astro_Exp\Radio 1\antdip_AE2025A.xlsx', sheet_name=0)
lndw = df['ln(dP)'].tolist()[0: -1]
ln_cosec_el = np.log(1/(np.sin(np.deg2rad(df["z"].tolist()))))[0: -1]

coeffs = np.polyfit(ln_cosec_el, lndw, deg=1)
fit = np.poly1d(coeffs)


plt.figure(figsize = (15, 15))
plt.plot(ln_cosec_el, fit(ln_cosec_el), color = 'orange', label = 'Fit', ls = '--')
plt.plot(ln_cosec_el, lndw, color = 'magenta', label = 'Unfit')
plt.legend()
plt.ylabel('$ln(\Delta w)$')
plt.xlabel('$ln(-sec(el))$')
plt.title('Unfitted data')
plt.show()

print(r'Negative Slope ($(\tau)_{0}$) = '+str(-coeffs[0]))
```

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

df = pd.read_excel(r'C:\Users\nicol\Documents\Astro_Exp\Radio 1\antdip_AE2025A.xlsx', sheet_name=0)
lndw = df['ln(dP)'].tolist()[0: -1]
ln_sec_el = np.log(np.abs(df["-SEC ('z')"].tolist()))[0: -1]

coeffs = np.polyfit(ln_sec_el, lndw, deg=1)
fit = np.poly1d(coeffs)


plt.figure(figsize = (15, 15))
plt.plot(ln_sec_el, fit(ln_sec_el), color = 'orange', label = 'Fit', ls = '--')
plt.scatter(ln_sec_el, lndw, color = 'magenta', label = 'Unfit')
plt.legend()
plt.ylabel('$ln(\Delta w)$')
plt.xlabel('$ln(-sec(el))$')
plt.title('Unfitted data')
plt.show()

print(r'Negative Slope ($(\tau)_{0}$) = '+str(-coeffs[0]))
```
